```python
Pipeline_id=''
Pipeline_Name=''
Pipeline_RunId=''
Dataset_Id=''
Dataset_Name=''
Start_Time=''
End_Time=''
Status=''
Error_Message=''
TruncateCurrentLogInd =''
InsertCurrentLogInd =''
InsertEXELogInd =''
InsertDTLogInd =''
UpdateCurrentLogInd =''
```

```python
from datetime import datetime

Error_Message=f"""{Error_Message}"""

if Start_Time==None or End_Time==None:
    Duration=None
else:
    def parse_timestamp(timestamp_str):
        parts = timestamp_str.split('.')
        dt = datetime.strptime(parts[0], "%Y-%m-%dT%H:%M:%S")
        microseconds = int(parts[1][:6])  # Truncate to microseconds
        return dt.replace(microsecond=microseconds)

    timestamp1 = parse_timestamp(Start_Time)
    timestamp2 = parse_timestamp(End_Time)

    # Find the difference between the timestamps
    time_difference = timestamp2 - timestamp1

    # Calculate total seconds in the time difference
    total_seconds = time_difference.total_seconds()

    # Calculate hours, minutes, and seconds from total seconds
    hours = int(total_seconds // 3600)
    minutes = int((total_seconds % 3600) // 60)
    seconds = int(total_seconds % 60)

    # Format the time difference as 'HH:mm:ss'
    Duration = "{:02d}:{:02d}:{:02d}".format(hours, minutes, seconds)
```

```python
if TruncateCurrentLogInd=='1':
    spark.sql("delete from pipeline_current_execution_log")
```

```python
if InsertCurrentLogInd=='1':
    if End_Time==None:
        spark.sql(f"insert into pipeline_current_execution_log (Pipeline_id,Pipeline_Name,Pipeline_RunId,Start_Time,End_Time,Duration,Status,Error_Message) values ({int(Pipeline_id)},'{Pipeline_Name}','{Pipeline_RunId}','{Start_Time}',NULL,'{Duration}','{Status}','{Error_Message}') ")
    else:
        spark.sql(f"insert into pipeline_current_execution_log (Pipeline_id,Pipeline_Name,Pipeline_RunId,Start_Time,End_Time,Duration,Status,Error_Message) values ({int(Pipeline_id)},'{Pipeline_Name}','{Pipeline_RunId}','{Start_Time}','{End_Time}','{Duration}','{Status}','{Error_Message}') ")
```

```python
if InsertEXELogInd=='1':
    if End_Time==None:
        spark.sql(f"insert into pipeline_history_execution_log (Pipeline_id,Pipeline_Name,Pipeline_RunId,Start_Time,End_Time,Duration,Status,Error_Message) values ({int(Pipeline_id)},'{Pipeline_Name}','{Pipeline_RunId}','{Start_Time}',NULL,'{Duration}','{Status}','{Error_Message}') ")
    else:
        spark.sql(f"insert into pipeline_history_execution_log (Pipeline_id,Pipeline_Name,Pipeline_RunId,Start_Time,End_Time,Duration,Status,Error_Message) values ({int(Pipeline_id)},'{Pipeline_Name}','{Pipeline_RunId}','{Start_Time}','{End_Time}','{Duration}','{Status}','{Error_Message}') ")
```

```python
if UpdateCurrentLogInd=='1':
    spark.sql(f"update pipeline_current_execution_log set Status='{Status}',End_Time='{End_Time}',Duration='{Duration}',Error_Message='{Error_Message}' where Pipeline_Name='PL_MASTER_E2E_EXECUTION' and Pipeline_RunId='{Pipeline_RunId}'")
```

```python
if InsertDTLogInd=='1':
    spark.sql(f"""insert into pipeline_dataset_logs (Pipeline_id,Pipeline_Name,Pipeline_RunId,Dataset_Id,Dataset_Name,Start_Time,End_Time,Duration,Status,Error_Message) values ({int(Pipeline_id)},'{Pipeline_Name}','{Pipeline_RunId}',{int(Dataset_Id)},'{Dataset_Name}','{Start_Time}','{End_Time}','{Duration}','{Status}',"{str(Error_Message)}") """)
```

 ## Create Raw Tables

```python
from notebookutils import mssparkutils
ls=mssparkutils.fs.ls('abfss://ed48e609-f66b-4cc4-832c-588ba37d0047@onelake.dfs.fabric.microsoft.com/e50527ec-11eb-4319-9a37-6a3ea5c7f29a/Files/source_file_temp')
for i in ls:
    df=spark.read.format('csv').option('header',True).load(f'Files/source_file_temp/{i.name}')
    df_col=df.columns
    for j in df_col:
        df=df.withColumnRenamed(j,j.replace(' ','_').lower())
    out=f"abfss://ed48e609-f66b-4cc4-832c-588ba37d0047@onelake.dfs.fabric.microsoft.com/e50527ec-11eb-4319-9a37-6a3ea5c7f29a/Files/source_file_temp/{i.name}.delta"
    df.write.format('delta').save(out)
```

## Create Error Logging Tables

```python
%%sql
drop table pipeline_history_execution_log
```

```python
%%sql
--Create Pipeline History Execution Log
create table pipeline_history_execution_log(
Pipeline_id int,
Pipeline_Name VARCHAR(50),
Pipeline_RunId VARCHAR(50),
Start_Time TIMESTAMP,
End_Time TIMESTAMP,
Duration VARCHAR(10),
Status VARCHAR(1),
Error_Message VARCHAR(1000))
```
